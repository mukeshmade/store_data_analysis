```python
orders_by_country = df.groupby("Country")["Quantity"].sum().reset_index().sort_values(by="Quantity", ascending=False)
highest_orders_country = orders_by_country.iloc[0]["Country"]
print(f"The country with the highest orders is {highest_orders_country}")
```

In this code, we first group the DataFrame by the "Country" column and sum the "Quantity" column for each country using the groupby function. We then sort the resulting DataFrame in descending order of "Quantity" and get the country with the highest orders from the first row of the sorted DataFrame.

```python
profit_by_category = df.groupby("Category")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False)
highest_profit_category = profit_by_category.iloc[0]["Category"]
lowest_profit_category = profit_by_category.iloc[-1]["Category"]
print(f"The category with the highest profit is {highest_profit_category}")
print(f"The category with the lowest profit is {lowest_profit_category}")
```

In this code, we first group the DataFrame by the "Country" column and sum the "Quantity" column for each country using the groupby function. We then sort the resulting DataFrame in descending order of "Quantity" and get the country with the highest orders from the first row of the sorted DataFrame.

### Business Insights
1. Sales and profit are highly correlated:
From the scatter plot and correlation matrix, we can observe that sales and profit are highly positively correlated. This means that as sales increase, profits also tend to increase. Therefore, the company should focus on increasing sales to increase profits.

2. Furniture category has the lowest profit margin:
From the bar chart of profit margin by category, we can observe that the furniture category has the lowest profit margin compared to the other two categories. This indicates that the company may need to re-evaluate the pricing strategy for furniture items or consider reducing the cost of production to improve the profit margin.

3. Profit margin is highest for the office supplies sub-category:
From the bar chart of profit margin by sub-category, we can observe that the office supplies sub-category has the highest profit margin compared to the other sub-categories. Therefore, the company should focus on increasing sales of office supplies to maximize profits.

4. Discounts negatively impact profit margin:
From the scatter plot and correlation matrix, we can observe that discounts have a negative correlation with profit margin. This means that as discounts increase, profit margin tends to decrease. Therefore, the company should consider reducing the frequency and amount of discounts to improve profit margins.

5. The company has a high presence in the western region:
From the bar chart of sales by region, we can observe that the western region has the highest sales compared to the other regions. This indicates that the company has a strong presence in the western region and should consider expanding to other regions to increase sales and profits.

6. Increase sales through online channels:
From the bar chart of sales by ship mode, we can observe that the online channel has lower sales compared to the other two channels. Therefore, the company should focus on increasing sales through the online channel by investing in online marketing and improving the user experience on the website.

7. Increase sales through targeted marketing:
From the bar chart of sales by state, we can observe that some states have significantly lower sales compared to others. The company should conduct targeted marketing campaigns in these states to increase sales and improve its market presence.

8. Optimize inventory management:
From the bar chart of sales by sub-category, we can observe that some sub-categories have significantly lower sales compared to others. The company should optimize its inventory management by reducing the stock of low-performing items and increasing the stock of high-performing items to maximize profits.

9. Improve customer retention:
From the scatter plot and correlation matrix, we can observe that discounts have a positive correlation with the number of items sold, but a negative correlation with profit margin. Therefore, instead of offering discounts to attract new customers, the company should focus on improving customer retention by offering loyalty programs and excellent customer service to retain existing customers and increase profits.

10. Monitor and analyze customer feedback:
The company should regularly monitor and analyze customer feedback to identify areas for improvement and address any negative feedback. This can be done through online surveys, social media monitoring, or customer service interactions. Improving customer satisfaction can lead to increased customer loyalty and ultimately, increased profits.

By implementing these recommendations, the company can improve its sales, profit margin, and customer satisfaction, which can lead to long-term growth and success.
